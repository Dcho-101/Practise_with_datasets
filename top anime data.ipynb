```python
# importing nesessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# importing warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Top_Anime_data.csv')
```

```python
# for EDA we don't need the description, japanese columns. 
# dropping those columns would be beeter for evaluation

df.drop(['Description', 'Japanese'], axis=1, inplace=True)
```

```python
# checking null values

df.isnull().sum()
```

```python
# dropping null rows

df.dropna(axis=0, inplace=True)
```

```python
df['Episodes'].value_counts().plot(figsize=(10,10),kind='barh')
```

```python
# there are several types of ratings available 

# ratings need to be cleaned for further processing

# checking unique values for rating

df['Rating'].unique()
```

```python
# there can be seen how the ratings are evaluated. 
# from these rating we can see the consumer for anime watcher

df['Rating'].value_counts().plot(kind='barh')
```

```python
# PG-13 - Teens 13 or older segments shows maximum consumers
```

```python
# Score Distribution
plt.figure(figsize=(10,5))
sns.histplot(df['Score'], bins=25, kde=True, color='r')
plt.title('Score Distribution')
plt.ylabel('Frequency')
plt.xlabel('Score')
plt.show()
```

```python
df.head()
```

```python
# Popularity

plt.figure(figsize=(10,8))
sns.histplot(df['Popularity'],bins=35,kde=True,color='g')
plt.title("Popularity Distribution")
plt.xlabel("Popularity")
plt.ylabel("Frequency")
plt.show()
```

```python
# Genre distribution (top 10)

# we need to split the genres

geners_split=df['Genres'].str.split(',').explode().str.strip()
plt.figure(figsize=(10,6))
sns.countplot(y=geners_split,order=geners_split.value_counts().index[:10])
plt.title("Top 10 Anime Genre")
plt.xlabel("Count")
plt.ylabel('Genre')
plt.show()
```

```python
# Top 10 Producers

top_producers = df['Producers'].str.split(',').explode().value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_producers.values,y=top_producers.index,)
plt.title("Top 10 Anime Producers")
plt.xlabel("Count")
plt.ylabel('Producers')
plt.show()
```
