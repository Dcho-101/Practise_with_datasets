```python
# Importing Libraries

import pandas as pd # data processing
import numpy as np # linear algebra
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Importing dataset

df = pd.read_csv('C:/Users/user/Downloads/archive (1)/apple_quality.csv')
```

```python
# Checking the dataset
df.head()
```

```python
# Checking the null values

df.isnull().sum()
```

```python
# cheking the dataset from last to check any null or missing values
df.tail()
```

### It can be seen that the dataset contains some null values at the end of the dataset. 
### This row does not store any data, so we can remove the row
### Also we can remove the 'A_id' as this column is not nessesary for the dataset

```python
df = df.drop('A_id', axis = 1)
df = df.drop(df.index[-1])
```

```python
df
```

```python
# Checking the types of the columns from the dataset 

df.info()
```

```python
# Checking null values

df.isnull().sum()
```

```python
## It seems that the dataset does not contain any null values now. but the dataset may have object datatype Acidity column.
## we need to convert the Acidity column to float datatypes for proper use of the dataset 
```

```python
# Changing the Acidity datatype into float

df['Acidity'] = df['Acidity'].astype(float)
```

```python
plt.figure(figsize = (8,5))
sns.countplot(data = df, x = 'Quality') # barplot cannot do x values as x is not numerical column
plt.title('Bar plot of Quality')
plt.show()
```

```python
# Size vs weight

plt.figure(figsize = (5,5), dpi = 150)
sns.scatterplot(data = df, x = 'Size', y = 'Weight', alpha = 0.5) # alpha for transparency
plt.show()
```

```python
# Size vs Sweetness
plt.figure(figsize=(5,5))
sns.scatterplot(data = df, x='Size', y = 'Sweetness', alpha = 0.5)
plt.show()
```

```python
# Size vs Crunchyness
plt.figure(figsize=(5,5), dpi = 150)
sns.scatterplot(data = df, x='Size', y = 'Crunchiness', alpha = 0.5)
plt.show()
```

```python
# size vs Acidity
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='Size',y='Acidity')
plt.show()
```

```python
quality_dummies = pd.get_dummies(df['Quality'], dtype='int')
quality_dummies
```

```python
# concatinating the quality_dummies dataframe
df = pd.concat([df,quality_dummies], axis = 1)
df.head()
```

```python
# droping the quality variable as it dont need now
df = df.drop(['Quality'], axis = 1)
df.head()
```

```python
#Correlation between features using heatmap

sns.heatmap(df.corr(),annot=True, cmap='Greens')
```
